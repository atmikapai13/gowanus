### map NYC BIDs using Folium library

```python
# Step 1: Import Libraries
# ------------------------
import pandas as pd
import geopandas as gpd
import folium
from shapely import wkt
```

```python
# Step 2: Explore NYC BIDs Dataset
# --------------------------------
print("Downloading NYC BIDs data...")
bids_url = "https://data.cityofnewyork.us/api/views/7jdm-inj8/rows.csv?date=20241117&accessType=DOWNLOAD"
bids_data = pd.read_csv(bids_url)

print("NYC BIDs Dataset Preview:")
bids_data.head()
```

```python
# use the the_geom column to create a GeoDataFrame. note that the_geom is a string column
bids_data["geometry"] = bids_data["the_geom"].apply(wkt.loads)
# check if there are missing geometries
missing_geometries = bids_data[bids_data["geometry"].isnull()]
if not missing_geometries.empty:
    print("Warning: Some geometries are missing.")
    display(missing_geometries)

bids_data.head()
```

```python
# Step 3: Map BIDs Using Folium
# -----------------------------
print("Creating a map of BIDs...")

# create a GeoDataFrame
bids_gdf = gpd.GeoDataFrame(bids_data, geometry="geometry")
# get the center of the map
center = bids_gdf["geometry"].unary_union.centroid.coords[0][::-1]
# create a map
m = folium.Map(location=center, zoom_start=11)
# add BIDs to the map
# ValueError: Cannot transform naive geometries.  Please set a crs on the object first.
bids_gdf.crs = "EPSG:4326"
bids_gdf.to_crs(epsg=4326)
folium.GeoJson(bids_gdf).add_to(m)
# show the map
m
```

```python
import plotly.express as px
import geopandas as gpd
from shapely import wkt
import json

print("Creating a map of BIDs...")

# Convert the_geom column to geometry objects
bids_data["geometry"] = bids_data["the_geom"].apply(wkt.loads)

# Create a GeoDataFrame
bids_gdf = gpd.GeoDataFrame(bids_data, geometry="geometry")
bids_gdf.set_crs(epsg=4326, inplace=True)

# Reproject to a suitable projected CRS (e.g., EPSG:3857 for Web Mercator)
bids_gdf_projected = bids_gdf.to_crs(epsg=3857)

# Calculate centroids in the projected CRS
centroids = bids_gdf_projected.geometry.centroid

# Convert centroids back to EPSG:4326 for mapping
centroids_4326 = centroids.to_crs(epsg=4326)

# Convert GeoDataFrame to GeoJSON
geojson = json.loads(bids_gdf.to_json())

# Create the map using Plotly Express
fig = px.choropleth_map(bids_gdf,
                           geojson=geojson,
                           locations=bids_gdf.index,
                           color=bids_gdf.index,
                           hover_name='F_ALL_BI_2',  # Change this to the column containing BID names
                           map_style="carto-positron",
                           zoom=12,
                           center={"lat": centroids_4326.y.mean(),
                                   "lon": centroids_4326.x.mean()},
                           opacity=1)

# Update layout for better visibility
fig.update_layout(
    title='Map of Business Improvement Districts (BIDs)',
    margin={"r":0,"t":50,"l":0,"b":0},
    height=800,
    width=1200
)

# Show the map
fig.show()
```

```python
# calculate the area in square foot of the BID from `geometry` column
bids_gdf["area_sqft"] = bids_gdf["geometry"].to_crs(epsg=2263).area

bids_gdf.head()
```

```python
# create a plotly plot where the x is the size of the BID and the y is the year of establishment from Year_Found column
import plotly.express as px
import plotly.graph_objects as go

fig = px.scatter(bids_gdf, y="area_sqft", x="Year_Found", title="BID Area vs. Year of Establishment",
                #  size of the dots by the area in square foot
                size="area_sqft",
                # color dots by the borough from F_ALL_BI_1 column
                color=bids_gdf["F_ALL_BI_1"],
                # hover data
                hover_name="F_ALL_BI_2",
                )

fig.show()
```

However, based on common naming conventions, it might represent:

So, abfe_0_2pc could represent the advisory base flood elevation for the 0.2% annual chance flood event.

```python
# Create the map using Plotly Express
fig = px.choropleth_map(floodplain_data,
                            geojson=floodplain_data.geometry,
                            locations=floodplain_data.index,
                            color=floodplain_data.index,
                            hover_name='fld_zone',  # Change this to the column containing flood zone names
                            map_style="carto-positron",
                            zoom=12,
                            center={"lat":
                                    # NYC's latitude
                                    40.730610,
                                      "lon":
                                    # NYC's longitude
                                    -73.935242},
                            opacity=0.5)

# Update layout for better visibility
fig.update_layout(
    title='Map of Business Improvement Districts (BIDs)',
    margin={"r":0,"t":50,"l":0,"b":0},
    height=800,
    width=1200
)

# Show the map
fig.show()
```

```python
import json
import plotly.express as px
import plotly.graph_objects as go

# Create a custom colorscale for BIDs (hues of pink)
bid_colorscale = px.colors.sequential.Reds[::-1]  # Reversed Reds colorscale for pink hues

# Create the map using Plotly Express
fig = px.choropleth_mapbox(bids_gdf,
                           geojson=geojson,
                           locations=bids_gdf.index,
                           color=bids_gdf.index,
                           color_continuous_scale=bid_colorscale,
                           hover_name='F_ALL_BI_2',  # Change this to the column containing BID names
                           mapbox_style="carto-positron",
                           zoom=12,
                           center={"lat": centroids_4326.y.mean(),
                                   "lon": centroids_4326.x.mean()},
                           opacity=0.7)

# Convert floodplain geometry to GeoJSON
floodplain_geojson = json.loads(floodplain_data.to_json())

# Create a custom colorscale for floodplain (darker blues)
flood_colorscale = [
    [0, "rgb(220,240,255)"],  # Light blue
    [0.5, "rgb(70,130,180)"], # Steel blue
    [1, "rgb(0,0,128)"]       # Navy
]

# Add floodplain data to the map
fig.add_choroplethmapbox(geojson=floodplain_geojson,
                         locations=floodplain_data.index,
                         z=floodplain_data.index,
                         colorscale=flood_colorscale,
                         showscale=False,
                         marker_opacity=0.5)

# Update layout for better visibility
fig.update_layout(
    title='Map of Business Improvement Districts (BIDs) and Floodplain Data',
    margin={"r":0,"t":50,"l":0,"b":0},
    height=800,
    width=1200
)

# Update color axis for BIDs
fig.update_coloraxes(colorbar_title="BIDs")

# Show the map
fig.show()
```
